```python
list3 = [14.6,17.8,24.2,30.6,37,31.1,25.2,19.2,13.3]
```

```python
list4=[]
list5=[]
list_cn = []

list6=[]
list7=[]

mu = 60/55
s = 5
N=10
for i in range(len(list3)):
    sum1=0
    sum2=0
    list_c=[]
    lamda = 352*0.96*0.2*list3[i]*0.01*(1/3)
    list4.append(round(lamda,2))
    list5.append(round(lamda/(s*mu),2))

    for n in range(1,s+1):
        list_c.append(((lamda/mu)**n)/math.factorial(n))
    for n in range(s+1,N+1):
        list_c.append(((lamda/mu)**n)/(math.factorial(s)*s**(n-s)))
    list_cn.append(list_c)
    
    #for n in range(1,s+1):
    #    sum1 += (lamda/mu)**n/math.factorial(n)+(lamda/mu)**s/math.factorial(s)
    #for n in range(s+1,N+1):
    #    sum2 += (lamda/(s*mu))**(n-s)
    #list6.append((1+sum1*sum2)**(-1)) #P_0

    list7.append(((1+sum1*sum2)**(-1))*(lamda/mu)**s*(lamda/(s*mu))/(math.factorial(s)*(1-lamda/(s*mu))**2)*(1-(lamda/(s*mu))**(N-s+1)-(N-s+1)*(lamda/(s*mu))**(N-s)*(1-lamda/(s*mu))))
```

```python
list_p0=[]
for i in range(len(list3)):
    list_p0.append(1/(1+sum(list_cn[i])))
print(list_p0) #P0 list
```

```python
list_pn=[]
for i in range(len(list3)):
    list_p=[]
    for n in range(N):
        list_p.append(list_cn[i][n]*list_p0[i])
    list_pn.append(list_p)
print(list_pn) #p1~p10
```
