```python
#train/test seperation
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df_main, test_size=0.2, random_state = 0)
validation_set, test_set = train_test_split(test_set, test_size=0.5, random_state = 0)

print('# of train_set : %.0f. # of validation_set : %.0f. # of test_set : %.0f' %(train_set.shape[0], validation_set.shape[0], test_set.shape[0]))
```

```python
from sklearn.utils import validation
#feature and label seperation
train_set_X = train_set.drop('SATURATOR_ML_SUPPLY_F_PV.Value', axis=1)
train_set_y = train_set['SATURATOR_ML_SUPPLY_F_PV.Value'].copy()

validation_set_X = validation_set.drop('SATURATOR_ML_SUPPLY_F_PV.Value', axis=1)
validation_y = validation_set['SATURATOR_ML_SUPPLY_F_PV.Value'].copy()

test_set_X = test_set.drop('SATURATOR_ML_SUPPLY_F_PV.Value', axis=1)
test_set_y = test_set['SATURATOR_ML_SUPPLY_F_PV.Value'].copy()
```

```python
train_set_y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_reg = LinearRegression()
lin_reg.fit(train_set_X, train_set_y)
print("lin_reg.coef_",lin_reg.coef_)
print("lin_reg.intercept_",lin_reg.intercept_)
```

```python
from sklearn.metrics import r2_score
final_predictions = lin_reg.predict(test_set_X)

final_mse = mean_squared_error(test_set_y, final_predictions)
final_rmse = np.sqrt(final_mse)

print('final_linear_RMSE : ', final_rmse)
print('Train_r2_score : ', lin_reg.score(train_set_X, train_set_y))
print('Test_r2_score : ', r2_score(test_set_y, final_predictions))
```

```python
from sklearn.linear_model import Ridge

# ridge regression (L2)
model_ridge = Ridge(alpha = 0.01)
model_ridge.fit(train_set_X, train_set_y)
print('Train_score : ', model_ridge.score(train_set_X, train_set_y))

# ridge RMSE
ridge_predicted = model_ridge.predict(test_set_X) #Test X
#y_hat
print('Test_score : ', r2_score(test_set_y, ridge_predicted))

delta = test_set_y - ridge_predicted
print('Errors in  prediction : ', np.mean(np.abs(delta)/test_set_y))

final_ridge_mse = mean_squared_error(test_set_y, ridge_predicted)
final_ridge_rmse = np.sqrt(final_ridge_mse)
print('final_ridge_RMSE : ', final_ridge_rmse)
```

```python
from sklearn.linear_model import Lasso

#lasso regression(L1)
model_lasso = Lasso(alpha = 10**-10)
model_lasso.fit(train_set_X, train_set_y)

print('Train_score : ', model_lasso.score(train_set_X, train_set_y))

# lasso RMSE
lasso_predicted = model_lasso.predict(test_set_X) #Test X
#y_hat
print('Test_score : ', r2_score(test_set_y, lasso_predicted))

delta = test_set_y - lasso_predicted
print('Errors in housing value prediction : ', np.mean(np.abs(delta)/test_set_y))

final_lasso_mse = mean_squared_error(test_set_y, lasso_predicted)
final_lasso_rmse = np.sqrt(final_lasso_mse)
print('final_ridge_RMSE : ', final_lasso_rmse)
```
