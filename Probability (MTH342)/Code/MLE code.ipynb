```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
```

```python
mu = 5
list1 = []
for i in np.linspace(1, 12, 100):
    sigma2 = i
    Product = 1
    for x in [1,4,5,6,9]:

        y = (1 / np.sqrt(2 * np.pi * sigma2)) * np.exp(-(x-mu)**2 / (2 * sigma2))

        Product = Product * y
       
    list1.append(Product)
    print(Product)
```

```python
plt.plot(np.linspace(1, 12, 100), list1, alpha=0.7, label='Likelihood function (mu=5)')
plt.xlabel('sigma_square')
plt.ylabel('Likelihood')
plt.legend(loc='lower right')
plt.axvline(6.8, 0, 0.97, color='red', linestyle='--', linewidth=2)
plt.show()
```

```python
sigma2 = 6.8
list2 = []
for i in np.linspace(1, 12, 100):
    mu = i
    Product = 1
    for x in [1,4,5,6,9]:

        y = (1 / np.sqrt(2 * np.pi * sigma2**2)) * np.exp(-(x-mu)**2 / (2 * sigma2**2))

        Product = Product * y
       
    list2.append(Product)
    print(Product)
```

```python
plt.plot(np.linspace(1, 12, 100), list2, alpha=0.7, label='Likelihood function (sigma_squre=6.8)')
plt.xlabel('mu (mean)')
plt.ylabel('Likelihood')
plt.legend(loc='lower left')
plt.axvline(5.0, 0, 0.97, color='red', linestyle='--', linewidth=2)
plt.show()
```
